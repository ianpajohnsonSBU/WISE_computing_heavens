# A Toy Model of the Expanding Universe

Now we go beyond Hubble’s original discovery and build a simple modern cosmology model.

Instead of just asking *“Is the universe expanding?”* we now ask:

- What is the universe made of?
- How fast is it expanding?
- Is the expansion slowing down or speeding up?

To answer this, we use a simplified version of modern cosmology called the **ΛCDM model**.

---

# The Expansion Rate

The expansion of the universe is described by the Hubble parameter:

$$
H(z)
$$

which tells us how fast the universe is expanding at a given redshift $z$.

We define a dimensionless function:

$$
E(z) = \frac{H(z)}{H_0}
$$

where:

- $H_0$ is the Hubble constant today  
- $z$ is redshift  

For a flat universe with matter and a cosmological constant:

$$
E(z) = \sqrt{\Omega_m (1+z)^3 + \Omega_\Lambda}
$$

---

# What Are the Omegas?

The symbols $ \Omega $ are **density parameters**.

They tell us what fraction of the universe’s total energy density comes from each component.

$$
\Omega = \frac{\rho}{\rho_{\text{critical}}}
$$

where

$$
\rho_{\text{critical}} = \frac{3H_0^2}{8\pi G}
$$

The critical density is the amount needed to make the universe spatially flat.

---

## Important Density Parameters

### $ \Omega_m $

Matter density:
- Includes normal matter (stars, gas)
- Includes dark matter
- Scales as $ (1+z)^3 $ because matter gets denser as we look back in time

### $ \Omega_\Lambda $

Cosmological constant (dark energy):
- Does **not** dilute as the universe expands
- Has constant energy density
- Causes accelerated expansion

---

# What Does "Flat" Mean?

A **flat universe** means the geometry of space follows Euclidean rules:

- Parallel lines stay parallel  
- Angles of a triangle add to 180°  
- The large-scale curvature of space is zero  

In cosmology, flatness means:

$$
\Omega_m + \Omega_\Lambda = 1
$$

There is no curvature term.

If:

- $ \Omega_m + \Omega_\Lambda > 1 $ → positively curved universe  
- $ \Omega_m + \Omega_\Lambda < 1 $ → negatively curved universe  
- $ \Omega_m + \Omega_\Lambda = 1 $ → flat universe  

In this lesson, we assume flatness, so:

$$
\Omega_m = 1 - \Omega_\Lambda
$$

This means if we measure $ \Omega_\Lambda $, we automatically know $ \Omega_m $.

---

# Luminosity Distance

To compare theory to supernova observations, we calculate the **luminosity distance**:

$$
D_L(z) = (1+z)\frac{c}{H_0} \int_0^z \frac{dz'}{E(z')}
$$

This accounts for:

- Expansion during light travel  
- Redshifting of photons  
- Geometry of space  

---

# Distance Modulus

Astronomers do not usually measure distance directly.

Instead, they measure brightness and compute the **distance modulus**:

$$
\mu = 5\log_{10}(D_L) + 25
$$

where:

- $D_L$ is in megaparsecs (Mpc)  
- $\mu$ is what supernova surveys report  

---

# Big Idea

By comparing:

- Observed supernova distance moduli  
- Predicted distance moduli from different values of $ \Omega_\Lambda $

we can determine whether the universe:

- Is slowing down (matter-dominated)  
- Or accelerating (dark-energy dominated)  

This is how late-20th-century supernova surveys discovered that the universe’s expansion is accelerating.
