```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
sun_radius_km = 696340  # Radius of the Sun in kilometers
G = 6.67430e-11  # Gravitational constant in m^3 kg^-1 s^-2
c = 299792458  # Speed of light in m/s
sun_mass_kg = 1.989e30  # Mass of the Sun in kilograms

# Calculate Schwarzschild radius (R_s = 2GM/c^2)
schwarzschild_radius_m = 2 * G * sun_mass_kg / c**2
schwarzschild_radius_km = schwarzschild_radius_m / 1000  # Convert to kilometers

# Generate circle points
theta = np.linspace(0, 2 * np.pi, 100000)
sun_x = sun_radius_km * np.cos(theta)
sun_y = sun_radius_km * np.sin(theta)

bh_x = schwarzschild_radius_km * np.cos(theta)
bh_y = schwarzschild_radius_km * np.sin(theta)

# Plot the circles
plt.figure(figsize=(8, 8))
plt.plot(sun_x, sun_y, label="Sun's Radius (696,340 km)", color='orange')
plt.plot(bh_x, bh_y, label=f"Schwarzschild Radius ({schwarzschild_radius_km:.2f} km)", color='black')

# Plot settings
plt.title("Sun and Schwarzschild Radius")
plt.xlabel("Distance (km)")
plt.ylabel("Distance (km)")
plt.legend()
plt.grid(True)

plt.show()
```
