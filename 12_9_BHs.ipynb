```python
import numpy as np
import matplotlib.pyplot as plt
from pycbc.waveform import get_td_waveform

# Parameters for the black holes
mass1 = 10  # Solar masses
mass2 = 10  # Solar masses
distance = 10  # Mpc

# Generate the waveform
hp, hc = get_td_waveform(approximant="SEOBNRv4_opt",
                         mass1=mass1,
                         mass2=mass2,
                         distance=distance,
                         delta_t=1.0/4096,
                         f_lower=20)
strain_array = hp.numpy()
times_array = hp.sample_times

# Plot the waveform
plt.plot(times_array, strain_array)
plt.xlabel('Time [s]')
plt.ylabel('Strain')
plt.title('Gravitational Waves from a Binary Black Hole Merger')
plt.grid(True)
plt.show()
```
